## Hypothesis


We predict a dual pattern between a country’s economic development and how students use ChatGPT for schoolwork. Specifically, we predict that higher-education students in more economically developed countries—measured using **GDP per capita (PPP)**—will report using ChatGPT **more often** for schoolwork. We also predict that, among students who use ChatGPT for schoolwork, students in **less developed (lower-income) countries** will use it relatively **more for assignment production** (drafting/writing/coding/generating deliverables), while students in **more developed (higher-income) countries** will use it relatively **more for learning support** (tutoring/explanations/summaries).

Our target group is **full-time undergraduate students** enrolled in **on-campus, in-person courses**, so we will narrow the dataset to only respondents who match this population.

We expect this pattern because access and adoption tend to be higher in wealthier countries, including more reliable internet access and fewer barriers to using ChatGPT (including language barriers). In lower-income settings, higher academic pressure and fewer support resources may push students toward using ChatGPT to produce deliverables. In contrast, in some higher-income settings, stricter academic integrity norms may discourage using ChatGPT to generate final assignments and instead encourage using it for learning support.

To examine this, we will use the self-reported survey data on (1) **how often students use ChatGPT for schoolwork** and (2) **what tasks they use it for**. For task type, we will group items into two categories. “Assignment production” will include tasks like **academic writing, professional writing, creative writing, and coding assistance**. “Learning support” will include tasks like **summarizing, calculations help, study assistance, and research assistance**. We will create a **Production score** (the average of production items) and a **Learning score** (the average of learning-support items). Then, among respondents who report using ChatGPT for schoolwork, we will label each student as **Production**, **Learning**, or **Mixed** based on which score is higher. 

We will then examine whether there is a relationship between a country’s **GDP per capita (PPP)** and (1) students’ **frequency of use** (Rarely, Occasionally, Moderately, Considerably, Extensively), and (2) their **task orientation**(Production vs Learning vs Mixed), based on whether their Production score or Learning score is higher. Our hypothesis will be supported if higher GDP per capita is associated with higher reported frequency and a more learning-support-oriented pattern, while lower GDP per capita is associated with lower frequency but a more production-oriented pattern.

### Dataset #1: Higher Education Students' Early Perceptions of ChatGPT

**Dataset Name:** Higher Education Students' Early Perceptions of ChatGPT: Global Survey Data

**Link to the dataset:** https://data.mendeley.com/datasets/ymg9nsn6kn/2

**Number of observations:** 23,218 higher-education students from 109 countries

**Number of variables:** 174 variables across 11 survey sections

**Description of the variables most relevant to this project:**

This dataset contains survey responses collected between October 2023 and February 2024 by the University of Ljubljana and 200+ partner institutions worldwide, capturing students' ChatGPT usage patterns during the first 12-15 months after the tool's release.

**Usage Frequency (Q15):** Students rated how often they use ChatGPT on a **5-point scale** (1=Rarely, 2=Occasionally, 3=Moderately, 4=Considerably, 5=Extensively). The **average score is 2.47**, meaning most users fall between "Occasionally" and "Moderately." This is a perception-based measure—students rate their usage relative to what they imagine others do, not in absolute terms like "sessions per week." **Values of 4-5 indicate heavy users**, while **1-2 represents light, infrequent use**.

**Usage Purpose (Q18 series):** Students rated how often they use ChatGPT for 12 different schoolwork tasks (1=Never to 5=Always). We group these into two categories:
- **Learning support** (Q18e: brainstorming, Q18g: summarizing, Q18i: study assistance, Q18k: research help): Tasks where ChatGPT aids understanding without generating final deliverables. **Average score: 2.60** ("Rarely" to "Sometimes").
- **Assignment production** (Q18a: academic writing, Q18h: calculating, Q18l: coding): Tasks where ChatGPT directly generates assignment content. **Average score: 2.25**, slightly lower than learning support.

From these, we calculate a **Learning Ratio** = [learning support score] / [learning + production scores], ranging from **0 (all production) to 1 (all learning)**. The **sample average is 0.51**, meaning students use ChatGPT roughly equally for both purposes, though substantial variation exists (SD=0.14).

**Country (Q4):** Students' study location, which we'll merge with external **World Bank GDP per capita data** to measure economic development. **Top countries:** Ecuador (8.2%), Spain (6.1%), Mexico (5.6%).

**Control Variables:** Individual **economic status (Q39)**: 1=significantly below average to 5=significantly above average, relative to peers in their country; **field of study (Q10)**; **academic level (Q8)**; **ChatGPT version (Q14)**: free vs. paid subscription.

**Descriptions of any shortcomings this dataset has with respect to the project:**

**Self-selection bias:** This survey was voluntary, so it **overrepresents students who are aware of ChatGPT** and interested enough to complete a 15-minute survey. Students who've never heard of ChatGPT, lack internet access, or are indifferent to AI are **systematically excluded**. This likely **inflates usage rates** compared to the true student population. However, if self-selection operates similarly across countries, **relative differences between wealthy and poor countries remain meaningful**.

**Uneven geographic coverage:** The **top 3 countries account for 20% of responses**, while **52 countries have fewer than 50 students**. Small-sample countries have **unreliable estimates**. We'll focus analysis on **countries with n>100**.

**Self-report bias:** All data is **self-reported**. Students may **misremember usage frequency** or **underreport using ChatGPT for "academic writing"** if they view it as cheating, while **overreporting "study assistance"** which sounds more legitimate. This could make students appear **more learning-oriented than they actually are**.

**Systematic missingness:** **31.4% of students have missing usage data** because they answered "No" to "Have you ever used ChatGPT?" and were routed past those questions. We'll analyze only the **68.6% who are users (n≈15,675 after cleaning)** when examining usage patterns.

**Temporal snapshot:** Data is from **2023-2024, when ChatGPT was new**. Usage patterns may have evolved since then as the technology and institutional responses matured.

### Filter Target Population

To ensure consistency with our research question, we restrict the dataset to the relevant student population.

### Create Composite Usage Scores

Next, we create two composite variables that summarize how students use ChatGPT for different academic purposes.

- **Learning Support Score**: average usage across tasks such as brainstorming, summarizing, studying, and research assistance.
- **Assignment Production Score**: average usage across tasks related to producing assignments such as academic writing, calculations, and coding.

These composite scores help simplify the analysis by grouping similar activities together and allow us to compare general learning support usage versus assignment production usage.

```python
learning_cols = ['use_brainstorm', 'use_summarize', 'use_study', 'use_research']
production_cols = ['use_academic', 'use_calculate', 'use_coding']

df_filtered['learning_score'] = df_filtered[learning_cols].mean(axis=1)
df_filtered['production_score'] = df_filtered[production_cols].mean(axis=1)
```

### Distribution of ChatGPT Usage Frequency

The first visualization examines the overall distribution of how frequently students in Dataset #1 report using ChatGPT for schoolwork. Students rated their usage on a 5-point scale ranging from "Rarely" to "Extensively."

This distribution is critical because usage frequency is one of our two main outcome variables. Understanding how usage is distributed helps us identify whether most students are light users, heavy users, or somewhere in between, and provides baseline context for comparing patterns across countries with different economic development levels.

```python
# Figure: Distribution of ChatGPT Usage Frequency
plt.figure(figsize=(10, 6))

freq_counts = df_filtered['use_frequency'].value_counts().sort_index()
plt.bar(freq_counts.index, freq_counts.values, color='steelblue', edgecolor='black', linewidth=1.5)

plt.xlabel('Usage Frequency', fontsize=12, fontweight='bold')
plt.ylabel('Number of Students', fontsize=12, fontweight='bold')
plt.title('Distribution of ChatGPT Usage Frequency for Schoolwork\n(Dataset #1: Early Perceptions 2023-2024)', 
          fontsize=13, fontweight='bold')
plt.xticks([1, 2, 3, 4, 5], 
           ['Rarely', 'Occasionally', 'Moderately', 'Considerably', 'Extensively'], 
           rotation=0)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

### Interpretation

The distribution shows that most students fall into the "Occasionally" to "Moderately" usage range, with a mean frequency of approximately 2.5. Heavy users (those reporting "Considerably" or "Extensively") represent a smaller proportion of the sample. This suggests that ChatGPT is used intermittently rather than as a primary tool for schoolwork among most students in this dataset.

### **GDP Analysis: Economic Development and ChatGPT Usage**

To answer our research question about whether GDP predicts ChatGPT usage, 
we now incorporate World Bank GDP per capita (PPP) data and examine the 
relationship between economic development and both usage frequency and 
learning orientation.

```python
###############################################################################
# GDP ANALYSIS FOR DATASET #1
# Add this code AFTER the "Cross-Country Variation" section in Dataset #1 EDA
###############################################################################

# ============================================================================
# STEP 1: RETRIEVE GDP PER CAPITA (PPP) DATA
# ============================================================================

print("\n" + "="*80)
print("Adding GDP Analysis to Dataset #1")
print("="*80)

import requests

indicator = "NY.GDP.PCAP.PP.CD"   # GDP per capita, PPP

url = f"https://api.worldbank.org/v2/country/all/indicator/{indicator}"
params = {
    "format": "json",
    "per_page": 20000
}

response = requests.get(url, params=params)
data = response.json()

records = []
for row in data[1]:
    if row["value"] is not None:
        records.append({
            "country_study": row["country"]["value"],
            "year": int(row["date"]),
            "gdp_per_capita_ppp": row["value"]
        })

gdp_pc_df = pd.DataFrame(records)

# Select most recent GDP value per country
gdp_pc_latest = (
    gdp_pc_df
    .sort_values(["country_study", "year"])
    .drop_duplicates(subset="country_study", keep="last")
)

print(f"Retrieved GDP data for {len(gdp_pc_latest)} countries")

# ============================================================================
# STEP 2: FIX COUNTRY NAME MISMATCHES
# ============================================================================

name_fixes = {
    'Bosnia & Herzegovina': 'Bosnia and Herzegovina',
    'Egypt': 'Egypt, Arab Rep.',
    'Gambia': 'Gambia, The',
    'Macedonia': 'North Macedonia',
    'Palestinian State*': 'West Bank and Gaza',
    'The Philippines': 'Philippines',
    'Turkey': 'Turkiye'
}

df_filtered['country_study'] = df_filtered['country_study'].replace(name_fixes)

# ============================================================================
# STEP 3: MERGE GDP DATA
# ============================================================================

df_filtered = df_filtered.merge(
    gdp_pc_latest[['country_study', 'gdp_per_capita_ppp']],
    on='country_study',
    how='left'
)

print(f"Successfully matched GDP data for {df_filtered['gdp_per_capita_ppp'].notna().sum()} students")

# ============================================================================
# STEP 4: CREATE COUNTRY-LEVEL SUMMARY WITH GDP
# ============================================================================

# Remove countries with missing GDP data
df_filtered_gdp = df_filtered[df_filtered['gdp_per_capita_ppp'].notna()].copy()

# Filter to countries with n >= 100 students
country_counts = df_filtered_gdp['country_study'].value_counts()
valid_countries = country_counts[country_counts >= 100].index
df_filtered_gdp = df_filtered_gdp[df_filtered_gdp['country_study'].isin(valid_countries)]

print(f"Analysis restricted to {len(valid_countries)} countries with n >= 100 students")
print(f"Final sample: {len(df_filtered_gdp)} students")

# Create country-level summary
country_summary = (
    df_filtered_gdp
    .groupby('country_study')
    .agg(
        n_students=('country_study', 'size'),
        mean_chatgpt_usage=('use_frequency', 'mean'),
        mean_learning_score=('learning_score', 'mean'),
        mean_production_score=('production_score', 'mean'),
        gdp_per_capita_ppp=('gdp_per_capita_ppp', 'first')
    )
    .reset_index()
)

# Calculate learning rate
country_summary['learning_rate'] = (
    country_summary['mean_learning_score'] / 
    (country_summary['mean_learning_score'] + country_summary['mean_production_score'])
)
country_summary['production_rate'] = 1 - country_summary['learning_rate']

# Create log GDP
country_summary['log_gdp'] = np.log(country_summary['gdp_per_capita_ppp'])

print("\nCountry-level summary created")
country_summary.head(10)

# ============================================================================
# VISUALIZATION 1: GDP DISTRIBUTION
# ============================================================================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Raw GDP
ax1.hist(country_summary['gdp_per_capita_ppp'], bins=20, color='steelblue', edgecolor='black')
ax1.set_xlabel('GDP per capita (PPP)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Number of Countries', fontsize=11, fontweight='bold')
ax1.set_title('Distribution of GDP per Capita (PPP)\n(Dataset #1: Early Perceptions 2023-2024)', 
              fontsize=12, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)

# Log GDP
ax2.hist(country_summary['log_gdp'], bins=20, color='coral', edgecolor='black')
ax2.set_xlabel('Log GDP per capita (PPP)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Number of Countries', fontsize=11, fontweight='bold')
ax2.set_title('Distribution of Log GDP per Capita\n(Dataset #1: Early Perceptions 2023-2024)', 
              fontsize=12, fontweight='bold')
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# INTERPRETATION
print("""
The GDP distribution is right-skewed, with most countries clustered in lower and middle 
income ranges. The log transformation creates a more balanced distribution that's easier 
to visualize and analyze.
""")

# ============================================================================
# VISUALIZATION 2: GDP vs CHATGPT USAGE FREQUENCY (WITH REGRESSION LINE)
# ============================================================================

plt.figure(figsize=(10, 7))

# Scatter plot
plt.scatter(country_summary['log_gdp'], 
            country_summary['mean_chatgpt_usage'],
            s=100, alpha=0.6, color='steelblue', edgecolor='black')

# Add country labels
for i, row in country_summary.iterrows():
    plt.text(row['log_gdp'], row['mean_chatgpt_usage'], 
             row['country_study'], fontsize=7, alpha=0.7)

# Add regression line
sns.regplot(x='log_gdp', y='mean_chatgpt_usage', 
            data=country_summary, scatter=False, 
            color='darkred', line_kws={'linewidth': 2})

plt.xlabel('Log GDP per capita (PPP)', fontsize=12, fontweight='bold')
plt.ylabel('Average ChatGPT Usage Frequency (1-5 scale)', fontsize=12, fontweight='bold')
plt.title('Relationship Between Economic Development and ChatGPT Usage Frequency\n(Dataset #1: Early Perceptions 2023-2024)', 
          fontsize=13, fontweight='bold')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# INTERPRETATION
print("""
INTERPRETATION:

The scatter plot reveals that there is **no clear linear relationship** between a 
country's GDP per capita and students' average ChatGPT usage frequency. The regression 
line is nearly horizontal, and data points are widely dispersed without a consistent 
upward or downward trend. This suggests that economic development alone does not 
systematically predict how often students use ChatGPT for schoolwork.

However, this does not mean GDP is irrelevant. Other factors may mediate or partition 
this relationship. For example, **usage patterns may differ by field of study** (STEM vs 
humanities), **institutional policies** (whether universities explicitly allow or ban 
ChatGPT), or **language background** (English-speaking vs non-English-speaking countries). 
Future analysis could partition countries by these characteristics to see if GDP 
relationships emerge within more homogeneous groups.
""")

# ============================================================================
# VISUALIZATION 3: GDP GROUPS vs USAGE FREQUENCY
# ============================================================================

# Create GDP groups
country_summary['gdp_group'] = pd.qcut(
    country_summary['log_gdp'], 3,
    labels=['Low GDP', 'Medium GDP', 'High GDP']
)

plt.figure(figsize=(8, 6))
sns.boxplot(x='gdp_group', y='mean_chatgpt_usage', data=country_summary,
            palette=['#ff9999', '#ffcc99', '#99ccff'])
plt.xlabel('GDP Group', fontsize=12, fontweight='bold')
plt.ylabel('Average ChatGPT Usage Frequency', fontsize=12, fontweight='bold')
plt.title('ChatGPT Usage Frequency Across GDP Groups\n(Dataset #1: Early Perceptions 2023-2024)', 
          fontsize=13, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# INTERPRETATION
print("""
The boxplot shows relatively similar median usage frequencies across GDP groups, with 
substantial overlap in the distributions. This further supports the finding that GDP 
alone does not strongly predict ChatGPT usage frequency in this dataset.
""")

# ============================================================================
# VISUALIZATION 4: GDP vs LEARNING ORIENTATION (WITH REGRESSION LINE)
# ============================================================================

plt.figure(figsize=(10, 7))

# Scatter plot
plt.scatter(country_summary['log_gdp'], 
            country_summary['learning_rate'],
            s=100, alpha=0.6, color='coral', edgecolor='black')

# Add country labels
for i, row in country_summary.iterrows():
    plt.text(row['log_gdp'], row['learning_rate'], 
             row['country_study'], fontsize=7, alpha=0.7)

# Add regression line
sns.regplot(x='log_gdp', y='learning_rate', 
            data=country_summary, scatter=False, 
            color='darkblue', line_kws={'linewidth': 2})

plt.xlabel('Log GDP per capita (PPP)', fontsize=12, fontweight='bold')
plt.ylabel('Share Using ChatGPT for Learning (vs Production)', fontsize=12, fontweight='bold')
plt.title('Relationship Between Economic Development and Learning-Oriented ChatGPT Use\n(Dataset #1: Early Perceptions 2023-2024)', 
          fontsize=13, fontweight='bold')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# INTERPRETATION
print("""
INTERPRETATION:

Similar to the usage frequency analysis, this scatter plot shows **no clear linear 
relationship** between GDP and the proportion of learning-oriented ChatGPT use. The 
regression line is nearly flat, and students in both high-income and low-income countries 
show comparable rates of using ChatGPT primarily for learning support versus assignment 
production.

This finding suggests that **economic development may not drive usage type** in a 
straightforward way. Instead, **other variables may be more important**. For instance:

- **Field of study**: Engineering students may use ChatGPT more for coding (production), 
  while social science students use it for summarizing (learning)
- **Academic pressure**: Countries with more competitive university systems may push 
  students toward production tasks regardless of GDP
- **Academic integrity norms**: Some high-GDP countries have stricter plagiarism policies, 
  potentially encouraging learning-oriented use

Future analysis should **partition results by these factors** to understand whether GDP 
relationships exist within specific student subgroups.
""")

# ============================================================================
# VISUALIZATION 5: GDP GROUPS vs LEARNING/PRODUCTION
# ============================================================================

task_usage = country_summary.melt(
    id_vars='gdp_group',
    value_vars=['learning_rate', 'production_rate'],
    var_name='Task_Type',
    value_name='Usage_Share'
)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Task_Type', y='Usage_Share', hue='gdp_group', data=task_usage,
            palette=['#ff9999', '#ffcc99', '#99ccff'])
plt.xlabel('ChatGPT Task Type', fontsize=12, fontweight='bold')
plt.ylabel('Share of ChatGPT Usage', fontsize=12, fontweight='bold')
plt.title('Learning vs Production ChatGPT Usage Across GDP Groups\n(Dataset #1: Early Perceptions 2023-2024)', 
          fontsize=13, fontweight='bold')
plt.xticks([0, 1], ['Learning Support', 'Assignment Production'])
plt.legend(title='GDP Group', fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================================
# VISUALIZATION 6: CORRELATION HEATMAP
# ============================================================================

corr_data = country_summary[['log_gdp', 'mean_chatgpt_usage', 'learning_rate', 'production_rate']]
corr_matrix = corr_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Between GDP and ChatGPT Usage Patterns\n(Dataset #1: Early Perceptions 2023-2024)', 
          fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# INTERPRETATION
print("""
The correlation heatmap confirms the weak relationships observed in the scatter plots. 
GDP shows minimal correlation with both usage frequency and learning orientation, 
suggesting that economic development alone does not predict how students use ChatGPT 
in this early adoption period (2023-2024).
""")

# ============================================================================
# SUMMARY STATISTICS
# ============================================================================

print("\n" + "="*80)
print("SUMMARY STATISTICS - DATASET #1 GDP ANALYSIS")
print("="*80)

print("\nCorrelation coefficients:")
print(f"  GDP vs Usage Frequency: r = {corr_matrix.loc['log_gdp', 'mean_chatgpt_usage']:.3f}")
print(f"  GDP vs Learning Rate: r = {corr_matrix.loc['log_gdp', 'learning_rate']:.3f}")
print(f"  GDP vs Production Rate: r = {corr_matrix.loc['log_gdp', 'production_rate']:.3f}")

print("\nMean usage by GDP group:")
gdp_group_summary = country_summary.groupby('gdp_group')['mean_chatgpt_usage'].mean()
for group, mean_usage in gdp_group_summary.items():
    print(f"  {group}: {mean_usage:.2f}")

print("\nMean learning rate by GDP group:")
learning_group_summary = country_summary.groupby('gdp_group')['learning_rate'].mean()
for group, mean_learning in learning_group_summary.items():
    print(f"  {group}: {mean_learning:.2%}")

print("\n" + "="*80)
print("GDP ANALYSIS FOR DATASET #1 COMPLETE!")
print("="*80)
```

## Cross-Dataset Comparison: Early (2023–2024) vs. Evolving (2024–2025) Perceptions

To examine whether ChatGPT usage patterns shifted over time, we compare key metrics
across Dataset #1 (Early Perceptions, 2023–2024) and Dataset #2 (Evolving Perceptions,
2024–2025). Both datasets use the same survey instrument, enabling direct comparison.

```python
# ── Align variable names across the two filtered datasets ──────────────────────

# Dataset #1 already has: use_frequency, learning_score, production_score
# Dataset #2 uses:        chatgpt_usage_extent, learning_score, production_score

ds1 = df_filtered[['use_frequency', 'learning_score', 'production_score']].copy()
ds1 = ds1.rename(columns={'use_frequency': 'usage_frequency'})
ds1['dataset'] = 'Dataset #1\n(Early 2023–24)'

ds2 = filtered_dataset_2[['chatgpt_usage_extent', 'learning_score', 'production_score']].copy()
ds2 = ds2.rename(columns={'chatgpt_usage_extent': 'usage_frequency'})
ds2['dataset'] = 'Dataset #2\n(Evolving 2024–25)'

combined = pd.concat([ds1, ds2], ignore_index=True)

print("Dataset #1 students:", len(ds1))
print("Dataset #2 students:", len(ds2))
print("\nMean usage frequency:")
print(combined.groupby('dataset')['usage_frequency'].mean())
print("\nMean learning score:")
print(combined.groupby('dataset')['learning_score'].mean())
print("\nMean production score:")
print(combined.groupby('dataset')['production_score'].mean())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=False)

for ax, (label, group) in zip(axes, combined.groupby('dataset')):
    freq_counts = group['usage_frequency'].value_counts().sort_index()
    ax.bar(freq_counts.index, freq_counts.values,
           color='steelblue', edgecolor='black', linewidth=1.2)
    ax.set_xticks([1, 2, 3, 4, 5])
    ax.set_xticklabels(['Rarely', 'Occasionally', 'Moderately', 'Considerably', 'Extensively'],
                       rotation=15, ha='right')
    ax.set_xlabel('Usage Frequency', fontsize=11, fontweight='bold')
    ax.set_ylabel('Number of Students', fontsize=11, fontweight='bold')
    ax.set_title(label, fontsize=12, fontweight='bold')
    ax.axvline(group['usage_frequency'].mean(), color='red', linestyle='--', linewidth=2,
               label=f'Mean: {group["usage_frequency"].mean():.2f}')
    ax.legend(fontsize=10)
    ax.grid(axis='y', alpha=0.3)

fig.suptitle('ChatGPT Usage Frequency: Dataset #1 vs. Dataset #2',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

Between the early (2023–24) and evolving (2024–25) survey waves, mean ChatGPT usage
frequency among students increased from 2.47 to 2.78 on a 5-point scale. The
distribution shifted rightward, with the modal response moving from "Occasionally"
to "Moderately." The share of students reporting rare use declined, while the
heavy-user tail ("Considerably" and "Extensively") grew. This suggests that ChatGPT
adoption among undergraduate students meaningfully increased over this one-year period,
consistent with broader trends of growing AI tool integration in academia.
